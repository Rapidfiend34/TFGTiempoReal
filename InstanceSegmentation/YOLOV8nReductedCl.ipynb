```python
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'TRUE'
import torch
from IPython.display import Image, clear_output
from ultralytics import YOLO

# No necesitas importar utils.downloads en YOLOv8

# Verificar configuración
print('Setup complete. Using torch %s %s' % (
    torch.__version__,
    torch.cuda.get_device_properties(0) if torch.cuda.is_available() else 'CPU'
))
from ultralytics import YOLO

# Cargar modelo
model = YOLO('yolov8n-seg.pt')

results = model.train(
    data='cityscapes.yaml',
    epochs=10,        # Aumentar épocas ya que la tendencia es positiva
    imgsz=640,
    batch=4,
    workers=0,

    # Mantener augmentaciones suaves pero ajustar algunas
    flipud=0.3,
    fliplr=0.5,
    mosaic=0.7,       # Reducir un poco más
    mixup=0.2,

    # Ajustar color ligeramente
    hsv_h=0.01,
    hsv_s=0.4,        # Reducir más la saturación
    hsv_v=0.3,

    # Mantener transformaciones geométricas suaves
    degrees=25.0,     # Reducir un poco más
    translate=0.1,
    scale=0.4,
    shear=5.0,
    perspective=0.0005,

    # Otras configuraciones
    copy_paste=0.2,
    augment=True,

    # Optimización
    warmup_epochs=2,
    warmup_momentum=0.8,
    lr0=0.001,        # Ajustar learning rate
    lrf=0.0001,       # Learning rate final
    weight_decay=0.0003  # Añadir regularización suave
)
```
