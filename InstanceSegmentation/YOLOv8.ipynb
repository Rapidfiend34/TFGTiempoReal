```python
import os
import json
import cv2
import numpy as np

# Rutas de Cityscapes
cityscapes_path = "data/cityscapes/"
gtFine_path = os.path.join(cityscapes_path, "gtFine", "train")
output_labels_path = "data/cityscapes/labels/train/"

# Crear carpeta de salida
os.makedirs(output_labels_path, exist_ok=True)

# Cargar clases de Cityscapes relevantes para segmentación de instancia
instance_classes = {"person": 0, "car": 1, "truck": 2, "bus": 3, "rider": 4, "bicycle": 5, "motorcycle": 6}

# Procesar cada archivo JSON de anotaciones
for city in os.listdir(gtFine_path):
    city_path = os.path.join(gtFine_path, city)
    for file in os.listdir(city_path):
        if file.endswith("_polygons.json"):
            json_path = os.path.join(city_path, file)
            with open(json_path, "r") as f:
                data = json.load(f)

            image_name = file.replace("_gtFine_polygons.json", "_leftImg8bit.png")

            txt_path = os.path.join(output_labels_path, image_name.replace(".png", ".txt"))
            
            with open(txt_path, "w") as label_file:
                for obj in data["objects"]:
                    label = obj["label"]
                    if label in instance_classes:
                        class_id = instance_classes[label]
                        polygon = np.array(obj["polygon"], dtype=np.float32)
                        polygon[:, 0] /= data["imgWidth"]  # Normalizar X
                        polygon[:, 1] /= data["imgHeight"]  # Normalizar Y
                        
                        # Convertir el polígono a cadena YOLO
                        polygon_str = " ".join(map(str, polygon.flatten()))
                        label_file.write(f"{class_id} {polygon_str}\n")
```
